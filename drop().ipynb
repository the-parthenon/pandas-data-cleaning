```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
```

```python
data.head()
```

#### If we use the drop() method with an index number, it drops that index/row

```python
data.drop(0)
```

#### To drop a column, switch to axis=1

```python
data.drop('id', axis=1)
```

#### Instead of axis, we can use the columns argument and pass in an array of column names to drop several at once.

```python
data.drop(columns=['height', 'width', 'depth'])
```

```python
data.head()
```

#### Similarly, we can drop multiple rows with the label argument

```python
data.drop(labels=[0, 1, 2])
```

#### To make actual changes to the dataframe, use inplace=True

```python
data.drop(columns=['id'], inplace=True)
```

```python
data.head()
```

#### We can import only the columns we want with the usecols argument

```python
data = pd.read_csv('artwork_sample.csv', usecols=['artist', 'title'])
data.head()
```
