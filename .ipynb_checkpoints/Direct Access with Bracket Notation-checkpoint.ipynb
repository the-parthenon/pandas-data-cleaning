```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
```

```python
data.head()
```

```python
data['id']
```

#### Square bracket notation returns a series, so we can use another square bracket to directly access the contents of a single value via index number

```python
data['id'][1]
```

#### We can pass in an array of column titles if we want to look at multiple columns at once

```python
data[['artist', 'title']]
```

#### Below, we get an error because the return is a df instead of a series

```python
data[['artist', 'title']][1]
```

```python
#data[1] will not return a row, but we can access a slice of rows using square brackets
data[1:2]
```

#### With a range of rows returning a df, we can access a column-name using square brackets

```python
data[1:5]['artist']
```

#### We can use filtering methods within the square brackets to get our range of rows instead of a slice.

```python
data[data['year'] > 1800]
```

#### Once again, we can add a second square bracket to get a series return of a single column.

```python
data[data['year'] > 1800]['year']
```
