```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_data.csv', low_memory=False)
```

```python
data.head()
```

#### In the dateText column, let's change "date not known" to NaN
First, we'll use `pd.isna()` to identify if there NaN exists in the column

```python
pd.isna(data.loc[:, 'dateText'])
```

```python
from numpy import nan
```

##### We can use the .replace() method to replace the text 'date not known' with NaN like so:

```python
data.replace({ 'dateText': { 'date not known': nan }})
```

#### `.replace` does not change the data inplace unless specified to, though, so use `inplace=True` when you are ready to commit the change

```python
data.replace({ 'dateText': { 'date not known': nan }}, inplace=True)
```

```python
data.head()
```

```python
data = pd.read_csv('artwork_data.csv', low_memory=False)
```

```python
data.head()
```

#### Alternately, we can use the `.loc` method to first filter the dataset, and then replace the desired values;   It is extremely important to include the `['dateText']` argument here, or we'll set everything in the df to NaN

```python
data.loc[data.dateText == 'date not known', ['dateText']] = nan
data.head()
```

#### Below, we filter the Year column for things that are not null and that contain anything other than a number.

```python
data.loc[data.year.notnull() & data.year.astype(str).str.contains('[^0-9]')]
```

#### Again, remember to set the column (`['year']`), in this case, or you will do *bad* things to your data

```python
data.loc[data.year.notnull() & data.year.astype(str).str.contains('[^0-9]'), ['year']] = nan
```
